```python
import pandas as pd
import chardet
```

```python
# List of file paths
file_paths = [
    "E:/Guvi/VS_code/Capstone_Project_2/Customers.csv",
    "E:/Guvi/VS_code/Capstone_Project_2/Exchange_Rates.csv",
    "E:/Guvi/VS_code/Capstone_Project_2/Products.csv",
    "E:/Guvi/VS_code/Capstone_Project_2/Sales.csv",
    "E:/Guvi/VS_code/Capstone_Project_2/Stores.csv"
]

# Dictionary to hold DataFrames
dataframes = {}

# Loop through each file path
for file in file_paths:
    with open(file, 'rb') as f:
        result = chardet.detect(f.read())
        encoding = result['encoding']

    # Name DataFrame after file
    df_name = file.split('/')[-1].replace('.csv', '')
    # Read the CSV with detected encoding
    dataframes[df_name] = pd.read_csv(file, encoding=encoding)
```

```python
# Calculate 'Delivery Time' as the difference between 'Delivery Date' and 'Order Date'
Sales_df['Delivery Time'] = (
    Sales_df['Delivery Date'] - Sales_df['Order Date']).dt.days

# Calculate the average delivery time from non-null delivery times
average_delivery_time = Sales_df['Delivery Time'].mean()

# Fill missing 'Delivery Date' values
Sales_df['Delivery Date'] = Sales_df['Delivery Date'].fillna(
    Sales_df['Order Date'] + pd.to_timedelta(average_delivery_time, unit='days'))

# Convert 'Delivery Date' to show only the date (YYYY-MM-DD)
Sales_df['Delivery Date'] = Sales_df['Delivery Date'].dt.date

# Drop the 'Delivery Time' column as it's no longer needed
Sales_df = Sales_df.drop(columns=['Delivery Time'])
```

```python
import mysql.connector
```

```python
# Connect to the database server
connection = mysql.connector.connect(
    host='localhost', user='root', password='12345', port=3306)

# Create a cursor object
cursor = connection.cursor()
```

```python
# Create the database
cursor.execute("CREATE DATABASE dataspark")
connection.commit()
```

```python
# Selecting the database
cursor.execute("USE dataspark")
connection.commit()
```

```python
# Create the Customers table
cursor.execute("create table Customers(CustomerKey int primary key, Gender varchar(50), Name varchar(100), City varchar(100), State_Code varchar(50), State varchar(100), Zip_Code varchar(50), Country varchar(100), Continent varchar(100), Birthday date);")
connection.commit()
```

```python
# Create the Exchange_Rates table
cursor.execute(
    "create table Exchange_Rates(Exchange_Date date, Currency char(3), Exchange float);")
connection.commit()
```

```python
# Create the Products table
cursor.execute("create table Products(ProductKey int primary key, Product_Name varchar(100), Brand varchar(100), Color varchar(100), Unit_Cost_USD float, Unit_Price_USD float, SubcategoryKey int, Subcategory varchar(100), CategoryKey int, Category varchar(100));")
connection.commit()
```

```python
# Create the Sales table
cursor.execute("create table Sales(Order_Number int, Line_Item int, Order_Date date, Delivery_Date date, CustomerKey int, StoreKey int, ProductKey int, Quantity int, Currency_Code char(3));")
connection.commit()
```

```python
# Create the Stores table
cursor.execute(
    "create table Stores(StoreKey int primary key, Country varchar(100), State varchar(100), Square_Meters float, Open_Date date);")
connection.commit()
```

```python
# Execute the SQL command to show tables
cursor.execute("show tables;")

# Iterate through the cursor and print each table name
print("List of tables in the database:")
for i in cursor:
    print(i)
```

```python
# Function to insert DataFrame into SQL table
def insert_dataframe_to_sql(df, table_name, connection):

    # Prepare the insert query based on the DataFrame's columns
    placeholders = ', '.join(['%s'] * len(df.columns))

    columns = ', '.join(df.columns)
    insert_query = f"INSERT INTO {
        table_name} ({columns}) VALUES ({placeholders})"

    # Convert DataFrame to list of tuples
    data = list(df.itertuples(index=False, name=None))

    # Insert data into the table
    with connection.cursor() as cursor:
        cursor.executemany(insert_query, data)

    # Commit the transaction to save the changes
    connection.commit()

    print(f"Data inserted into {table_name} successfully.")
```

```python
insert_dataframe_to_sql(Customers_df, 'customers', connection)
```

```python
insert_dataframe_to_sql(ExchangeRate_df, 'exchange_rates', connection)
```

```python
insert_dataframe_to_sql(Products_df, 'products', connection)
```

```python
insert_dataframe_to_sql(Sales_df, 'sales', connection)
```

```python
insert_dataframe_to_sql(Stores_df, 'stores', connection)
```
